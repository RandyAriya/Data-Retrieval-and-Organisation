```python
import random    
import time
import pandas as pd

operators = ["+", "-", "*", "/"]
question_list = {'Question':[], 'Correct':[], 'Operator':[], 'Time':[]}

time_limit=30

#adjust the values within random.randint() to adjust difficulty
#operators can be targeted by running the loops within each operator and adjusting the range

score=0
time_finish = time.time() + time_limit
while time.time() < time_finish:
    i = random.choice(operators)
    if i == "+":
        for question_a in range (1):
            int_a_1 = random.randint(1, 250)
            int_a_2 = random.randint(1, 250)
            operator = "+"
            question_a = str(int_a_1) + i + str(int_a_2)
            q_t0 = time.time()           
            answer = input(question_a)
            if answer == str(eval(question_a)):
                question_list['Time'].append(time.time()-q_t0)                
                question_list['Question'].append(question_a)
                question_list['Correct'].append("Y")
                question_list['Operator'].append(i)
                score+=1
            else:
                print("wrong")
                question_list['Time'].append(time.time()-q_t0)                
                question_list['Question'].append(question_a)
                question_list['Correct'].append("N")
                question_list['Operator'].append(i)                
    elif i == "-":
        for question_s in range (1):
            operator_s = "-"
            int_s_1 = random.randint(2, 250)
            int_s_2 = random.randint(1, int_s_1)
            question_s = str(int_s_1) + i + str(int_s_2)
            q_t0 = time.time()
            answer = input(question_s)
            if answer == str(eval(question_s)):
                question_list['Time'].append(time.time()-q_t0)                
                question_list['Question'].append(question_s)
                question_list['Correct'].append("Y")
                question_list['Operator'].append(i)
                score+=1
            else:
                print("wrong")
                question_list['Time'].append(time.time()-q_t0)                
                question_list['Question'].append(question_s)
                question_list['Correct'].append("N")
                question_list['Operator'].append(i)  
    elif i == "*":
        for question_m in range (1):
            int_m_1 = random.randint(1, 12)
            int_m_2 = random.randint(1, 12)
            operator_m = "*"
            question_m = str(int_m_1) + i + str(int_m_2)
            q_t0 = time.time()            
            answer = input(question_m)
            if answer == str(eval(question_m)):
                question_list['Time'].append(time.time()-q_t0)                
                question_list['Question'].append(question_m)
                question_list['Correct'].append("Y")
                question_list['Operator'].append(i)
                score+=1
            else:
                print("wrong")
                question_list['Time'].append(time.time()-q_t0)                
                question_list['Question'].append(question_m)
                question_list['Correct'].append("N")
                question_list['Operator'].append(i)  
    elif i == "/":
        for question_d in range(1):
            operator_d = "/"
            int_d_1 = random.randint(2, 250)
            factors = []
            for p in  range(2, int_d_1+1):
                if int_d_1%p == 0: 
                    factors.append(str(p))
            if len(factors) == 1: 
                continue
            question_d = str(int_d_1) + i + str(random.choice(factors[:-1]))
            q_t0 = time.time()            
            input_answer = input(question_d)
            if input_answer == str(int(eval(question_d))):
                question_list['Time'].append(time.time()-q_t0)                
                question_list['Question'].append(question_d)
                question_list['Correct'].append("Y")
                question_list['Operator'].append(i)
                score+=1
            else:
                print("wrong")
                question_list['Time'].append(time.time()-q_t0)                
                question_list['Question'].append(question_d)
                question_list['Correct'].append("N")
                question_list['Operator'].append(i) 
print("Score: " + str(score))

df = pd.DataFrame(question_list)
df['Time'].describe()
```
